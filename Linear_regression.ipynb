##**Linear Regression**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salary_data = pd.read_csv('Salary_Data.csv')
X = salary_data.iloc[:,:-1].values
Y = salary_data.iloc[:,1].values
```

```python
# ploating the data

sns.displot(salary_data['YearsExperience'],kde=False,bins = 5);
```

```python
sns.countplot(y='YearsExperience',data=salary_data);
```

```python
sns.barplot(x='YearsExperience',y='Salary',data=salary_data);
```

```python
# Started with thhe model testing and training using machine learning
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 1/3,random_state = 0)
```

```python
# Fitting the Linear regression to the training set
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# Predicting the test result
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
# Visualising the Training set of result
```

```python
plt.scatter(X_train,y_train,color = 'blue')
plt.plot(X_train,lr.predict(X_train),color = 'red')
plt.title('Salary - Experience (Train set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visulising the testing result
```

```python
plt.scatter(X_test,y_test,color = 'blue')
plt.plot(X_test,lr.predict(X_test),color = 'red')
plt.title('Salary - Experience (Test set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Calculating the error or Resudials (THEY ARE INVERSLY PROPOTIONAL TO THE ACCURACY)
```

```python
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print ('RMSE:',np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
